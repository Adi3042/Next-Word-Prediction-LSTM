```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
# Fit tokenizer on full text (learns vocabulary)
tokenizer.fit_on_texts([faqs])
```

```python
tokenizer.word_index
```

```python
len(tokenizer.word_index)
```

```python
# Convert full text into sequence of numbers
tokenizer.texts_to_sequences([faqs])   # 2D list → use [0] for 1D
```

```python
# List to store all input sequences
input_sequences = []
for sentence in faqs.split('\n'):
  # Convert each sentence into token sequence
  tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

  # Create n-gram sequences
  for i in range(1,len(tokenized_sentence)):
    input_sequences.append(tokenized_sentence[:i+1])    
    # Example:
        # My
        # My Name
        # My Name is
        # My Name is Aditya
```

```python
input_sequences
```

```python
# Find max sequence length
max_len = max([len(seq) for seq in input_sequences])
vocab_size = len(tokenizer.word_index) + 1  # Vocabulary size (+1 because tokenizer index starts from 1)
```

```python
tokenizer.word_index
```

```python
# Pad Sequences
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Pad all sequences to same length, pre==># pad at beginning
padded_input_sequences = pad_sequences(input_sequences, maxlen = max_len, padding='pre')
```

```python
padded_input_sequences
```

```python
# Features: all words except last
X = padded_input_sequences[:, :-1]
```

```python
# Labels: last word of each sequence
y = padded_input_sequences[:, -1]
```

```python
# Shape of X (samples, sequence_length)
X.shape
```

```python
# Shape of y (samples,)
y.shape
```

```python
# Convert y to categorical (one-hot encoding)
from tensorflow.keras.utils import to_categorical
y = to_categorical(y, num_classes=vocab_size)
```

```python
# Shape after one-hot encoding
y.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
from tensorflow.keras.layers import Dropout
```

```python
model = Sequential()

# Embedding layer converts word index → dense vector
# vocab_size ==> total vocabulary
# 100 ==> embedding dimension
model.add(Embedding(vocab_size, 100, input_length=max_len-1))

# First LSTM layer (returns full sequence)
model.add(LSTM(150, return_sequences=True))   

# Dropout to reduce overfitting
model.add(Dropout(0.2))

# Second LSTM layer
model.add(LSTM(150))   

# Output layer: predict next word
model.add(Dense(vocab_size, activation='softmax'))
```

```python
# Compile model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
# Train the model
history = model.fit(X, y, epochs=200)
```
